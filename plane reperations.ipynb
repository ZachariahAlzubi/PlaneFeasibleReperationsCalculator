```python
# Suppose each passenger has the same probability p of showing up for the N seats on a flight.
# Each empty seat represents a loss of L dollars. Let m be the number of extra tickets sold.


import math

def binomial_prob(N, p, y):
    # calculates the binomial probability
    return math.comb(N, y) * (p ** y) * ((1 - p) ** (N - y))

def expected_value(N, p):
    # calculates the expected value of the number of empty seats
    ev = 0
    i=0
    for y in range(N + 1):
        bp = binomial_prob(N, p, y)
        print(i, " people show up:",bp*100,"%")
        ev += y * bp
        i+=1
        print ("expected value: ",ev)
        print()
    return ev

def max_compensation(N, p, L, m):
    # calculates the maximum compensation offered to each passenger
    ev = expected_value(N, p)
    R = (ev * L) / m
    return R

# example values for N, p, L, and m
N = 100
p = 0.95575
L = 100
m = 5

result = max_compensation(N, p, L, m)
print("The maximum compensation to offer is:", result)



# This code calculates the maximum compensation offered to each passenger to minimize the expected loss due to 
# empty seats on a flight. The input parameters are:

# N: the number of seats on the flight
# p: the probability that each passenger has of showing up for their seat
# L: the loss incurred by the airline for each empty seat
# m: the number of extra tickets sold
# The code uses the binomial distribution to model the number of empty seats on a flight and calculates the 
# expected value of the number of empty seats. The expected value is then used to calculate the maximum compensation 
# offered to each passenger.

# The main functions in this code are:

# binomial_prob(N, p, y): calculates the binomial probability of y passengers showing up for N seats, 
#     given that each passenger has a probability of p of showing up.
# expected_value(N, p): calculates the expected value of the number of empty seats on the flight, given N seats and a
#     probability of p for each passenger to show up.
# max_compensation(N, p, L, m): calculates the maximum compensation offered to each passenger to minimize the 
#     expected loss due to empty seats on a flight. The expected loss is calculated as the expected value of the 
#     number of empty seats multiplied by the loss per empty seat, L, and divided by the number of extra tickets sold, m.
# Finally, the code uses example values for N, p, L, and m, and calculates the maximum compensation to offer using 
# the max_compensation function. The result is then printed to the console.
```

```python
import math

def binomial_prob(N, p, y):
    # calculates the binomial probability
    return math.comb(N, y) * (p ** y) * ((1 - p) ** (N - y))

def expected_value(N, p):
    # calculates the expected value of the number of empty seats
    ev = 0
    for y in range(N + 1):
        ev += y * binomial_prob(N, p, y)
    return ev

def max_compensation(N, p, L, m):
    # calculates the maximum compensation offered to each passenger
    ev = expected_value(N, p)
    R = (ev * L) / m
    return R

# example values for L and m
L = 100
m = 5

# loop over a range of probabilities (90% to 100%) and oversold seats (1 to 10 seats)
compensation = []
for N in range(90, 101):
    for oversold in range(1, 11):
        p = N / 100
        result = max_compensation(N + oversold, p, L, m)
        compensation.append(result)

# calculate the average compensation
avg_compensation = sum(compensation) / len(compensation)
print("The average compensation to offer is:", avg_compensation)
```
